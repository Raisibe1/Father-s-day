```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv('marks.csv', sep=';')
df.set_index('Module', inplace=True)

print("DataFrame after setting 'Module' as index:\n", df)


num_modules = df.shape[0]
fig, axes = plt.subplots(num_modules, 1, figsize=(8, 3 * num_modules))


if num_modules == 1:
    axes = [axes]

for ax, (module_name, data) in zip(axes, df.iterrows()):
    data.plot(kind='bar', ax=ax, title=f'Marks for {module_name}')
    ax.set_xlabel('Assessment')
    ax.set_ylabel('Mark')
    ax.grid(True)

plt.tight_layout()
plt.show()


top3 = df.sort_values(by='Module Mark', ascending=False).head(3)

plt.figure(figsize=(8, 5))
top3['Module Mark'].plot(kind='bar', color='green')
plt.title('Top 3 Modules by Final Mark')
plt.xlabel('Module')
plt.ylabel('Module Mark')
plt.grid(axis='y')
plt.show()
```
