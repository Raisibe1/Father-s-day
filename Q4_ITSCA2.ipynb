```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

data = pd.read_csv('electricity_bill.csv', sep=';')
data.columns = ['Units_Consumed', 'Bill_Amount']
print(data.head())

plt.figure(figsize=(8, 5))
plt.scatter(data['Units_Consumed'], data['Bill_Amount'], color='blue', alpha=0.5, label='Data Points')
plt.title('Relationship Between Units Consumed and Electricity Bill')
plt.xlabel('Units Consumed')
plt.ylabel('Bill Amount')
plt.grid(True)
plt.legend()
plt.show()

X = data[['Units_Consumed']]
y = data['Bill_Amount']

model = LinearRegression()
model.fit(X, y)

plt.figure(figsize=(8, 5))
plt.scatter(data['Units_Consumed'], data['Bill_Amount'], color='blue', alpha=0.5, label='Data Points')
plt.plot(data['Units_Consumed'], model.predict(X), color='red', label='Best Fit Line')
plt.title('Units Consumed vs. Electricity Bill (With Regression Line)')
plt.xlabel('Units Consumed')
plt.ylabel('Bill Amount')
plt.grid(True)
plt.legend()
plt.show()

slope = model.coef_[0]
intercept = model.intercept_
print("\nSlope (Coefficient): {:.4f}".format(slope))
print("Y-Intercept: {:.4f}".format(intercept))

user_units = float(input("\nEnter the number of units consumed: "))
predicted_bill = model.predict([[user_units]])[0]
print(f"Predicted Electricity Bill for {user_units} units: R{predicted_bill:.2f}")

```
